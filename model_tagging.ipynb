## Creating a Speaker Feature:

```python
#setting n/a as default, so i can easily tell who has been updated below.
combined_transcripts['Speaker'] = 'n/a'
```

I know that if I split each transcript item by the ":" character, any first item of that split (0 index) that is all caps is definitely a candidate name.  This won't get everything, but will be a good start.

```python
for i, text in enumerate(combined_transcripts['Transcript']):
    split_text = text.split(':')
    if split_text[0].isupper():
        combined_transcripts.iloc[i, 5] = split_text[0]
```

Now, I'll need to see how I can fill in more of these n/a's.  One thought is to loop through each Debate - if the first line of the debate as a speaker, I can then loop through add in new speakers.

```python
debates = set(combined_transcripts.Debate_Name.tolist())
```

```python
for debate in debates:
    #First, call up the Dataframe with only that debate listed:
    new_df = combined_transcripts[combined_transcripts.Debate_Name == debate].reset_index()
    #Next, I need to check if the first row has a speaker listed, if not this won't work.
    if new_df.Speaker.iloc[0] != 'n/a':
        #Now, loop through the speakers.  If the current row does not have a speaker, add the previous row's speaker.
        for i, row in enumerate(new_df['Speaker']):
            if row == 'n/a':
                new_df.Speaker.iloc[i] = new_df.Speaker.iloc[i-1]
                #Use this value to reassign the value in the combined_transcripts frame:
                combined_transcripts.Speaker.iloc[int(new_df['index'].iloc[i])] = str(new_df.Speaker.iloc[i-1])
```
